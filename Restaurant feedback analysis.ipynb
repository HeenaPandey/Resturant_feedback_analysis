### Checking the duplicate data in the dataset

```python
print(f'Duplicates in restaurant dataset: {data.iloc[:,1:].duplicated().sum()}')
percentage_duplicated = np.round(100 * data.iloc[:, 1:].duplicated().sum() / len(data), 1)
print(f"{percentage_duplicated}%")
```

When we tried to check for the duplicated values in the dataset, no duplicacy was observed. So we can conclude that there are no duplicate values in the dataset.

```python
# Drop duplicates
data=data.drop_duplicates(subset=data.columns[1:], keep='first')
```
